# **5. Data Pre-Profiling**

This section is emphasised on getting a report about the data (Use Pandas Profiling)
For quick analysis pandas profiling is very handy.
Generates profile reports from a pandas DataFrame.
For each column, statistics are presented in an interactive HTML report.

```
profile = pandas_profiling.ProfileReport(df=df_az)
profile.to_file(output_file='Pre Profiling Report.html')
print('Accomplished!')
```

- The report shows that there are a total of **6 features** out of which **1** are **categorical**, **3** are **numerical**, and **2 unsupported** type.

- Features such as **uniq_ID** and **Product Name** are found to be **highly cardinal** in nature.

- The **Price** is **highly correlated** with the **rating** and vice-versa.

- Features such as **uniq_ID** and **Product Name,manufacturer, price,numberes  of answered questions and rating**  are found to be uniformly distributed.

- Features such as **uniq_ID and Product Name,manufacturer, price,numberes  of answered questions and rating** are found to have unique values.

- **uniq_ID and Product Name** are the unsupported type of feature. They needs cleaning or further analysis.

- **Note**: Rest of the information can be analyzed from the profile report itself.

### **Popular Manufacturers**

Q1. Who is the most popular manufacturer of Amazon?

```


%matplotlib inline
df_az["manufacturer"].value_counts().plot(kind="bar", color='g')
plt.xlabel("Manufacturers")
plt.ylabel("Products")
plt.show()
```

Q2. What are the average prices of Amazon?

```
%matplotlib inline
fig = plt.Figure()
plt.hist(x=df_az['price'], bins=20)
plt.xticks(np.arange(0,250,24))
plt.tick_params(axis='x', which='major', labelsize=6.5)
plt.xlabel("Price")
plt.ylabel("No. of products")
plt.show()
```

Q3. How many products received rating of 5 stars?

```
%matplotlib inline
star5 = 0
plt.hist(df_az['rating'], color='r', bins=12)
plt.xticks(np.arange(3.5,5.1,0.1))
plt.yticks(np.arange(0,650,50))
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.title("Ratings")

for x in df_az['rating']:
  if x == 5:
    star5+=1 


plt.show()
```

Q4. Are expensive products worth the price?

```
plt.scatter(x=df_az['price'], y=df_az['rating'], color='orange')
plt.xlabel('Price')
plt.ylabel('Rating')
plt.show()
```

Q5. How many questions are generally answered?

```
plt.hist(df_az['number_of_answered_questions'], color='c')
plt.xticks(np.arange(0,12,1))
plt.show()
```

Q6. What is the relation between brands and other values?

```
sns.set_style("whitegrid")
sns.pairplot(df_az, hue='manufacturer')
plt.show()
```
